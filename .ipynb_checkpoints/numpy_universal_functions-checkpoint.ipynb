# butterfly curve

```python
t = np.linspace(0, 12*np.pi, 360*5)
t
```

```python
x = np.sin(t)*(np.exp(np.cos(t))-2*np.cos(4*t)-np.sin(t/12)**5)
```

```python
y = np.cos(t)*(np.exp(np.cos(t))-2*np.cos(4*t)-np.sin(t/12)**5)
```

```python
plt.plot(x,y)
```
