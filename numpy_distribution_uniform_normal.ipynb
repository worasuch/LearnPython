```python
g=np.array(['rock', 'peper', 'scissors'], dtype='<U8')
g
```

```python
pet=np.array(['cat', 'dog', 'bird', 'fish'])
c=np.random.choice(pet, size=100, replace=True)
c
```

```python
u, counts=np.unique(c, return_counts=True)
print(u)
print(counts)
```

```python
plt.bar(u, counts)
```

```python
pet=np.array(['cat', 'dog', 'bird', 'fish'])
c=np.random.choice(pet, size=100, replace=True, p=[.2, .4, .1, .3])
c
```

```python
u, counts=np.unique(c, return_counts=True)
print(u)
print(counts)
```
